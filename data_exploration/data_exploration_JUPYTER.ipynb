```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
sns.set(style='whitegrid')
```

## Loading the Dataset

```
import os
# os.chdir('path/to/your/folder')

data = pd.read_csv('social_media_user_behavior.csv')
data.head()
```

## Clean Column Names
Standardize column names for easier handling.

```
data.columns = (
    data.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
)
data.columns
```

## Initial Data Exploration

```
data.info()
data.describe(include='all')
```

```
data.shape, len(data), data.shape[1]
```

```
data.head(10)
```

## Explore Categorical Variables

```
data['primary_platform'].value_counts()
data['gender'].value_counts()
```

## Explore Numeric Variables

```
data['age'].describe()
data['daily_usage_hours'].describe()
```

## Quick Visualization

```
sns.countplot(data=data, x='primary_platform')
plt.title('User Distribution by Platform')
plt.xticks(rotation=45)
plt.show()
```

## Data Selection & Filtering

```
data[['user_id', 'age', 'primary_platform']]
```

```
data[data['age'] > 25]
```

```
data[(data['age'] > 25) & (data['primary_platform'] == 'Instagram')]
```

## Data Manipulation

```
data['age_group'] = np.where(data['age'] < 25, 'Young', 'Adult')
```

```
data['usage_level'] = np.select(
    [data['daily_usage_hours'] > 3, data['daily_usage_hours'] > 1],
    ['High', 'Medium'],
    default='Low'
)
```

```
data = data.rename(columns={'age': 'user_age', 'primary_platform': 'platform_name'})
```

## Handling Missing Data

```
data.isna().sum()
```

```
data['user_age'] = data['user_age'].fillna(data['user_age'].mean())
```

## Data Aggregation & Summarization

```
data.groupby('platform_name').agg(avg_time=('daily_usage_hours','mean'), total_users=('user_id','count')).reset_index()
```

## String Cleaning with Regex

```
data['email'].str.contains(r'@gmail\.com', na=False)
```

```
data['username_clean'] = data['username'].str.replace('@','', regex=False)
```

## Reshaping Data

```
data_long = pd.melt(data, id_vars=['user_id'], var_name='variable', value_name='value')
data_long.head()
```

## Joining Datasets

```
user_info = pd.DataFrame({'user_id':['USR00001','USR00002'],'country':['UK','USA']})
pd.merge(data, user_info, on='user_id', how='left')
```

## Final Visualization

```
sns.scatterplot(data=data, x='user_age', y='daily_usage_hours')
sns.regplot(data=data, x='user_age', y='daily_usage_hours', scatter=False)
plt.show()
```
